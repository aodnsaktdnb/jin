## 집합자료형 Set
집합자료형의 특징<br/>
- 중복을 허용하지 않는다.<br/>
- 순서가 없다.<br/>
- 필터로 사용가능.<br/>
- 리스트나 튜플로 사용하면 요소에 접근가능.<br/>

```python
arr = [0,1,0,7,7,6,6,1,2,3,4]
print (arr)
s1 = set(arr)
print(s1)
```

```python
#교집합 구하기
s1 = set ([1,2,3,1,4,4,5,6])
print('s1: ',s1)
s2 = set ([3,4,3,5,5,6,7,7])
print('s2: ',s2)
s3 = s1 & s2
# s1.intersection(s2)
print('교집합: ',s3)

#합집함 구하기
s4 = s1 | s2
# s1.union(s2)
print('합집합 :', s4)

#차집합 구하기 
s5 = s1 - s2
s6 = s2 - s1
print('차집합1: ',s5)
print('차집합2: ',s6)
```

```python
# 값 1개 추가하기 - add
st1 = set([1,1,2,2,3,3])
st1.add(4)
print(st1)
```

```python
# 값 여러개 추가하기 - update
st2 = set([1,1,2,2,3,3])
st2.update([4,5,6,4,5,6])
print(st2)
```

```python
파이썬에서 깊은 복사와 얕은 복사는 객체의 특성에 따라 다르게 작동합니다. 

    1. 리스트 (List):
       - 깊은 복사**를 하면 새로운 메모리 공간에 리스트의 모든 요소를 복사합니다. 따라서 원본 리스트와 복사된 리스트는 서로 다른 메모리 주소를 가지게 되어 `id` 값이 달라집니다¹².

    2. **문자열 (String)과 튜플 (Tuple):
       - 문자열과 튜플은 **불변 객체 (immutable objects)**입니다. 즉, 한 번 생성되면 그 값을 변경할 수 없습니다. 
       - 깊은 복사를 하더라도 새로운 객체를 생성하지 않고, 기존 객체를 그대로 참조합니다. 따라서 id 값이 동일하게 유지됩니다²³.

    이러한 차이는 파이썬의 메모리 관리 방식과 객체의 가변성(mutable) 또는 불변성(immutable)에 기인합니다.
```

### 변수를 만드는 다양한 방법

```python
# 튜플 
a1 , b1 = 'python', 'life'
a2 , b2 = ('python', 'life')
(a3 , b3) = ('python', 'life')
(a4,b4,c4,d4) = 'life'
print(a1, b1)
print(a2, b2)
print(a3, b3)
print(a4,b4,c4,d4)
```

```python
#리스트
c1, d1 = ['python','life']
[c2,d2] = ['python','life']
[c3,d3] = ['python','life']
c4,d4 = ['python','life']
(c5,d5) = ['python','life']
print(c1,d1)
print(c2,d2)
print(c3,d3)
print(c4,d4)
print(c5,d5)
```
