```python
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.font_manager as fm

plt.style.use('fivethirtyeight')
d2coding_path = './D2Coding-Ver1.3.2-20180524.ttf'
fm.fontManager.addfont(d2coding_path)
matplotlib.rcParams["font.family"] = 'D2coding'

#마이너스 부로가 깨지지 않도록 설절 
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import geopandas as gpd
gdf = gpd.read_file('./data_set/LARD_ADM_SECT_SGG_26_202405.shp')
print(gdf)
#컬러캡 설정 
fig, ax = plt.subplots(1, 1, figsize=(20,20))
gdf.plot(column = 'SGG_NM', ax=ax, legend=True, cmap='tab20')

#지도그리기
plt.show()
```
