```python
from hdfs import InsecureClient
import pandas as pd
import folium
```

```python
dfs = InsecureClient('http://namenode:9870', user='hadoop')
path = "/user/hadoop/starbucks_20241227151418.csv"
with dfs.read(path,encoding='utf-8') as reader:
    data = pd.read_csv(reader)
data.columns = ["no.","name","lat","lon","addr","type"]
data.head()
```

```python
data = data[['name','lat','lon']]
data.head()
```

```python
# 데이터 준비
coordinates = data[["lat","lon"]].to_numpy()
place = data["name"].to_numpy()

# 중심 좌표 설정
center_lat, center_lon = 35.17574571 , 129.0817409
m = folium.Map(location=[center_lat,center_lon], zoom_start= 11)

# 좌표마킹
for i, (lat, lon) in enumerate(coordinates):
    folium.CircleMarker(
        location=[lat,lon],
        radius= 10,
        color="blue",
        fill=True,
        fill_color="skyblue",
        fill_opacity=0.5,
        tooltip=place[i]
    ).add_to(m)
# 라벨 추가 (Marker로 위치 조정)
    folium.Marker(
        location=[lat + 0.001, lon],  # 위도 약간 증가
        icon=folium.DivIcon(html=f'<div style="font-size: 9px; border-radius:5px; color: black; background-color: white; width:50px;white-space: nowrap; overflow: hidden;text-overflow: ellipsis;">{place[i]}</div>')
    ).add_to(m)
m
```
